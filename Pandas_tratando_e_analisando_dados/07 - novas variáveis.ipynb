## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

### criando a variável "Valor Bruto"

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(5)
```

### criando a variável "Valor m2"

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(5)
```

#### arredondando o numero de casas decimais do "Valor m2"

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head(5)
```

### criando a variável "Valor Bruto m2"

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(5)
```

## Criando variável agregada para diferenciar casas de apartamentos  

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```

## Excluindo Variáveis

### maneiras de excluir variáveis de um DataFrame: 
#### DataFrame_exemplo

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux.head(5)
```

```python
# maneira_1: del
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(5)
```

```python
# maneira_2: pop()
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux.head(5)
```

## Excluindo as variáveis: "Valor Bruto" e "Valor Bruto m2" do DataFrame "dados".
### utilizando o método 'drop()'.

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

```python
dados.head(10)
```

### exportando e subscrevendo o arquivo "aluguel_residencial.csv"

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```

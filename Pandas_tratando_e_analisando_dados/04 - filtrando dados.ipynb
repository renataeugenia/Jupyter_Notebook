### Exercicios

Veja o código que cria um DataFrame de letras e números.


```python
import pandas as pd

numeros = [i for i in range(11)]
letras = [chr(i + 65) for i in range(11)]
nome_coluna = ['N']

df = pd.DataFrame(data = numeros, index = letras, columns = nome_coluna)
```

```python
df
```

E baseado nesses dataframe geramos uma selecao para filtrar os dados:

```python
selecao = df['N'].isin([i for i in range(11) if i % 2 == 0])
df = df[selecao]
df
```

 O código acima cria um DataFrame com uma coluna chamada N e atribui os números de 0 até 10. Com o método isin(), selecionamos somente os números pares de N.

## extra_aula 04 - organizando dataframes (Sort)

```python
data = [[1,2,3], [4,5,6], [7,8,9]]
```

```python
data
```

```python
list('321')
```

```python
df1 = pd.DataFrame(data, list('321'), list('ZYX'))
df1
```

```python
# organiza pelo indice das linhas
df1.sort_index(inplace=True)
df1
```

```python
# organiza pelo indice das colunas
df1.sort_index(inplace=True, axis=1)
df1
```

```python
# organiza sorteando pela variavel escolhida da coluna (by='')
df1.sort_values(by='X', inplace=True)
df1
```

```python
# organiza sorteando pela variavel escolhida da linha (by='')
df1.sort_values(by='3', axis=1, inplace=True)
df1
```

```python
# é possivel passar uma lista de variaveis (by='')
df1.sort_values(by=['X','Y'], inplace=True)
df1
```
